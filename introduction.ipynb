````python
class CollegeMem:
    def __init__(self,id,name,age):
        self.id=id
        self.name=name
        self.age=age
    def display(self):
        print(f"College Id: {self.id},\nName:{self.name}\nAge:{self.age}")
        
class Teacher(CollegeMem):
    def __init__(self,id,name,age,subject, salary):
        CollegeMem.__init__(self,id,name,age)
        self.sub=subject
        self.sal=salary
    def display(self):
        #CollegeMem.display(self)
        super().display()
        print(f"Subject: {self.sub},\nSalary: {self.sal}")
        
tobj=Teacher(10121,'Vidya',34,"Python",55888)
tobj.display()
````
